3.Linear Regression
Building a simple linear Model where:

dependent variable=Monthly  charges

independent variable=tenure

Steps:

. Dividing datasets into 70:30 train-test split

. Building model on train set and predict onn test set

. Finding root mean square error

. Finding the error in prediction and store the result in error

. Finding root mean square error again


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
X = df[['tenure']]
y = df['MonthlyCharges']
```

Spliting the dataset (70% training and 30% testing)

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3, random_state = 42)
```

Model training

```python
lr=LinearRegression()
lr.fit(X_train,y_train)
```

Modeling Testing

```python
y_pred = lr.predict(X_test)
```

```python
error = y_test - y_pred
```

```python
rmse = np.sqrt(mean_squared_error(y_test,y_pred))
```

Building a simple Logistic regression model:

Dependent variable = Churn

independent variable = MonnthlyCharges

65:35 train-test split

Build connfusion matrix and calculate accuaracy score

build a muliple logistic regression model

dependent variable = churn

independent variable = tenure and monthly charges

build confusion matrix and calculate accuracy score

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
df['Churn']
```

Converting yes or no to 1 and 0. Because whenever we pass column the string format is not allowed instead it allows numeric format

```python
df['Churn'] = df['Churn'].map({'Yes':1,'No':0})
df['Churn']
```

```python
X = df[['MonthlyCharges']]
y = df['Churn']
```

Spliting the Dataset(65% train and 35% test)

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.35, random_state = 42)
```

Training the model

```python
log_reg = LogisticRegression()
log_reg.fit(X_train,y_train)
```

```python
y_pred = log_reg.predict(X_test)
```

```python
#To check accuracy we use confusion matrix

conf_matrix = confusion_matrix(y_test,y_pred)
conf_matrix
```

```python
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

Classification report detail breakdown of the model performance

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
